```python
import pypdf
import pandas as pd
```

```python
reader = pypdf.PdfReader('Objetos Conhecimento.pdf')

KC_RAW = ''
for page in reader.pages:
    KC_RAW = KC_RAW+str(page.extract_text())
KC_RAW
```

```python
KC_RAW_MATH = KC_RAW.split('2. Matemática e suas Tecnologias')[1].split('3. Ciências da Natureza e suas Tecnologias')[0]

lista_kc_matematica = []
for sub_topico in KC_RAW_MATH.split('•'):
    if ':' in sub_topico:
        sub_topico_nome = sub_topico.split(':')[0]
        for item_kc in sub_topico.split(':')[1].split(';'):
            lista_kc_matematica.append(sub_topico_nome+' - '+item_kc.replace('\n',''))
        
lista_kc_matematica
```

```python
KC_RAW_FISICS = KC_RAW.split('3. Ciências da Natureza e suas Tecnologias')[1].split('4. Ciências Humanas e suas Tecnologias')[0]

lista_kc_fisica = []
KC_RAW_FISICS  = KC_RAW_FISICS.split('3.1 Física')[1].split('3.2 Química')[0]
for sub_topico in KC_RAW_FISICS.split('•'):
    if ':' in sub_topico:
        try:
            sub_topico_nome = sub_topico.split('-')[0]
            for item_kc in sub_topico.split('-')[1].split('.'):
                if len(item_kc)>5:
                    lista_kc_fisica.append(sub_topico_nome+' - '+item_kc.replace('\n',''))
        except:
            sub_topico_nome = sub_topico.split('–')[0]
            for item_kc in sub_topico.split('–')[1].split('.'):
                if len(item_kc)>5:
                    lista_kc_fisica.append(sub_topico_nome+' - '+item_kc.replace('\n',''))
            
lista_kc_fisica
```

```python
KC_RAW_CHEM = KC_RAW.split('3. Ciências da Natureza e suas Tecnologias')[1].split('4. Ciências Humanas e suas Tecnologias')[0]

lista_kc_quimica = []
KC_RAW_CHEM  = KC_RAW_CHEM.split('3.2 Química')[1].split('3.3 Biologia')[0]
for sub_topico in KC_RAW_CHEM.split('•'):
    if ':' in sub_topico:
        try:
            sub_topico_nome = sub_topico.split('-')[0]
            for item_kc in sub_topico.split('-')[1].split('.'):
                if len(item_kc)>5:
                    lista_kc_quimica.append(sub_topico_nome+' - '+item_kc.replace('\n',''))
        except:
            sub_topico_nome = sub_topico.split('–')[0]
            for item_kc in sub_topico.split('–')[1].split('.'):
                if len(item_kc)>5:
                    lista_kc_quimica.append(sub_topico_nome+' - '+item_kc.replace('\n',''))
            
lista_kc_quimica
```

```python
aux_mat  = pd.DataFrame()
aux_mat['KC'] = lista_kc_matematica
aux_mat['disciplina'] = 'Matematica'

aux_fsc  = pd.DataFrame()
aux_fsc['KC'] = lista_kc_fisica
aux_fsc['disciplina'] = 'Fisica'

aux_qmk  = pd.DataFrame()
aux_qmk['KC'] = lista_kc_quimica
aux_qmk['disciplina'] = 'Quimica'

df = pd.concat([aux_mat,aux_fsc,aux_qmk])
df
```

```python
df.to_parquet('lista_kcs.parquet')
```
